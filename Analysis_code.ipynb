### Analyzing each column and checking weather it has any discrepency in dataset and fixing it.

```python
#### > "Age" Column

data['Age'].value_counts()
```

```python
## Cleaning above discrepency

data['Gender'] = [str(gender).replace('Femal','Female').replace('Femle','Female').replace('Femalee','Female') for gender in data['Gender']]
```

```python
df1[df1['Partner_salary'].isnull()].index
```

```python
for i in df1[df1['Partner_salary'].isnull()].index:
    if df1['Partner_working'][i] == 'No':
        df1['Partner_salary'][i] = 0
    elif df1['Partner_working'][i] == 'Yes':
        df1['Partner_salary'][i] = df1['Total_salary'][i] - df1['Salary'][i]
```
